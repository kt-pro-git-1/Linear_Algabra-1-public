## [3]　平面と直線の方程式


####  直線の方程式  (2次元)　平行なベクトルを使うとき


直線を通る１つの固定点をA$（A_x,A_y）$とし、直線に平行な１つの方向ベクトルを$ \boldsymbol{a}=(a_x,a_y) $とする

直線上を移動する任意の点を$P(x,y)$とすると

$\vec{AP}=(x , y)-(A_x , A_y)= (x - A_x,y - A_y)$

t: 実数とする　ベクトル$\vec{AP}と\boldsymbol{a}$が<font color="Red">平行な条件</font>は
$\vec{AP}=t \boldsymbol{a}$である

$ (x,y) -(A_x , A_y) = t (a_x,a_y )$　または　$ (x,y) =(A_x , A_y) + t (a_x,a_y )$

$(x ,y)=(A_x+ta_x,A_y+ta_y ) $より

$x =A_x+ta_x,  y= A_y+ta_y  $  ,または  $x -A_x = ta_x,  y- A_y =ta_y  $ 

$t$で共通にしてみると

$$t= \frac{ x - A_x}{a_x}=  \frac{y - A_y}{a_y}$$

上式より　$$ y= \frac{a_y}{a_x}(x-A_x)+A_y $$ とも表される

###### 垂直なベクトルを使うとき   直線の方程式  (2次元)


直線を通る１つの固定点をA$（A_x,A_y）$とし、直線に垂直な１つの方向ベクトルを$ \boldsymbol{n}=(n_x,n_y) $とする

直線上を移動する任意の点を$P(x,y)$とすると

$\vec{AP}=(x , y)-(A_x , A_y)= (x - A_x,y - A_y)$

<font color="Red">直交条件</font>より

$\vec{AP}\cdot \boldsymbol{n} =0 $である


$(x - A_x,y - A_y)\cdot(n_x,n_y)=0$

$$ n_x(x - A_x)+n_y(y - A_y) =0 $$

または、$ n_y \neq 0 $のとき　$$ y = \left(-\frac{n_x}{n_y}\right)x + \left(\frac{n_x}{n_y}A_x + A_y\right) $$

##  直線の方程式  (3次元　空間)


直線を通る１つの固定点をA$（A_x,A_y,A_z）$とし、直線に平行な１つの方向ベクトルを（$ \boldsymbol{a}=(a_x,a_y,a_z) $　）とする

直線上を移動する任意の点を$P(x,y,z)$とすると

$\vec{AP}=(x,y,z)-(A_x , A_y,A_z)= (x - A_x,y - A_y.z-A_z)$

t: 実数とする　ベクトル$\vec{AP}と\boldsymbol{a}$が<font color="Red">平行な条件</font>は
$\vec{AP}=t \boldsymbol{a}$である

$ (x,y,z)-(A_x , A_y,A_z) = t (a_x,a_y,a_z)$　または　$ (x,y,z)=(A_x , A_y,A_z) + t (a_x,a_y,a_z)$

$(x ,y)=(A_x+ta_x,A_y+ta_y ,A_z+ta_z) $より

$x =A_x+ta_x,  y= A_y+ta_y , z= A_z+ta_z  $  ,または  $x -A_x = ta_x,  y- A_y =ta_y ,  z- A_z =ta_z  $ 

$t$で共通にしてみると

$$t= \frac{x - A_x}{a_x}=  \frac{y - A_y}{a_y} =  \frac{z - A_z}{a_z}$$

と表される


### 直線の方程式

### $$ (x,y,z)=(A_x , A_y,A_z) + t (a_x,a_y,a_z) \tag{1.8}$$

### $$ \frac{x - A_x}{a_x}=  \frac{y - A_y}{a_y} =  \frac{z - A_z}{a_z} \tag{1.9}$$




直線の方程式  (2次元)  上にあるものと同じ

##  平面の方程式 （3次元　空間）



平面に固定する１つの点A$（A_x,A_y,A_z）$とし、平面に垂直な１つの方向ベクトルを$ \boldsymbol{n}=(n_x,n_y,n_z) $とする

平面上を移動する任意の点を$P(x,y,z)$とすると

$\vec{AP}=(x,y,z)-(A_x , A_y,A_z)= (x - A_x,y - A_y.z-A_z)$

内積の<font color="Red">直交条件</font>より

$ \vec{AP}\cdot \boldsymbol{n} = 0 $ (実数でのゼロ)　より

$ (x - A_x,y - A_y.z-A_z)\cdot (n_x,n_y,n_z) =0 $

### $$ n_x(x - A_x)+n_y(y - A_y)+n_z(z-A_z)=0  \tag{1.7}$$

原点(0,0,0)を通る場合には、
$ ax+by+cz =0$　の形になり

一般的な平面の方程式は、
$ ax+by+cz =d$　の形になる
